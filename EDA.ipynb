```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/cdmbi/FPD/master/data/FPD-non-redundant-processed.csv")
```

```python
data.tail(10)
```

```python
df = data[['excitation_new', 'emission_new', 'emission_color_class']]
```

```python
df.head(4)
```

```python
g = sns.PairGrid(df, hue = "emission_color_class")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
